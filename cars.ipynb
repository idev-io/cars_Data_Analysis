```python
# Import Required Packages
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
# Load datasets 
dataset_path = r'datasets/cars.csv'
cars = pd.read_csv(dataset_path)
```

```python
# Dataset exploration
cars.head()
```

```python
# Dataset Info
cars.info()
```

```python
# Dataset Basic Statistics
cars.describe()
```

```python
# Check Dataset Null Values
cars.isnull().sum()
```

```python
# Check shape of dataset
cars.shape
```

```python
# Drop 'Unnamed: 5' and 'RATING' columns
cars.drop(['Unnamed: 5', 'RATING'], axis = 1, inplace = True) 
```

```python
# Data Visualization
# Total CITY (kWh/100 km) for each Model
model_grp = cars.groupby('Model')['CITY (kWh/100 km)'].agg('sum').to_frame()
model_grp.style.background_gradient(cmap = 'Reds')
```

```python
fig, ax = plt.subplots(figsize = (15, 8))
sns.barplot('Make', '(kW)', data = cars, ax = ax, ci = False)
plt.show()
```

```python
sns.set_style('whitegrid')
fig, ax = plt.subplots(figsize = (15, 8))
sns.regplot('CITY (kWh/100 km)', '(km)', data = cars)
plt.show()
```

```python
fig, ax = plt.subplots(figsize = (15, 8))
sns.swarmplot('Model', 'CITY (kWh/100 km)', data = cars, ax = ax)
plt.xticks(rotation = 90)
plt.show()
```

```python
fig, ax = plt.subplots(figsize = (15, 8))
sns.boxplot('Size', 'HWY (kWh/100 km)', data = cars, ax = ax)
plt.show()
```

```python
fig, ax = plt.subplots(figsize = (15, 8))
sns.countplot('Model', data = cars, ax = ax)
plt.xticks(rotation = 90)
plt.show()
```

```python
cross_cars = pd.crosstab(cars['Model'], cars['Size'],
                         values = cars['(kW)'], aggfunc = 'mean')

fig, ax = plt.subplots(figsize = (15, 8))
sns.heatmap(cross_cars, annot = True, linewidths = 0.5, fmt = 'f')
plt.show()
```
